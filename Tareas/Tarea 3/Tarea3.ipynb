Se tiene un archivo (dates.txt) donde cada línea de este corresponde a una nota médica y cada nota tiene una fecha que debe extraerse, pero cada fecha está codificada en uno de muchos formatos. Por ejemplo, se muestra a continuación una lista de algunas de las variantes que se puede encontrar en este conjunto de datos:
<ul>
    <li>04/20/2009; 04/20/09; 4/20/09; 4/3/09</li>
    <li>Mar-20-2009; Mar 20, 2009; March 20, 2009; Mar. 20, 2009; Mar 20 2009;</li>
    <li>20 Mar 2009; 20 March 2009; 20 Mar. 2009; 20 March, 2009</li>
    <li>Mar 20th, 2009; Mar 21st, 2009; Mar 22nd, 2009</li>
    <li>Feb 2009; Sep 2009; Oct 2010</li>
    <li>6/2008; 12/2009</li>
    <li>2009; 2010</li>
</ul>
La actividad consiste en:
<p>a). Identificar correctamente todas las diferentes variantes de fecha codificadas en este conjunto de datos, normalizar y ordenar adecuadamente las fechas.</p>
<p>b).	Una vez que haya extraído estos patrones de fecha del texto, el siguiente paso es clasificarlos en orden cronológico ascendente de acuerdo con las siguientes reglas:</p>
<ul>
    <li>Todas las fechas están en formato xx/xx/xx son mm/dd/aa</li>
    <li>Todas las fechas en las que el año está codificado en solo dos dígitos corresponden a años posteriores a la década de 1900 (p. Ej., 1/5/89 es el 5 de enero de 1989).</li>
    <li>Si falta el día (p. Ej., 9/2009), suponga que es el primer día del mes (p. Ej., septiembre, 1 de 2009).</li>
    <li>Si falta el mes (por ejemplo, 2010), suponga que es el primero de enero de ese año (p. Ej., enero, 1 de 2010).</li>
    <li>Tenga cuidado con los posibles errores tipográficos, ya que este es un conjunto de datos derivados de la vida real.</li>
</ul>
Esta función debería devolver una lista de longitud 500.

```python
# Vamos expotar  las librerias nnecesarias:
import pandas as pd
import re
```

```python
# Vamos a abrir y leer los datos de forma que  ada renglon sea un registro de una 
# lista, para poder iterar despues.
with open('dates.txt') as f:
    #contents = f.readlines()
    contents = f.readlines()
```

```python
def fechas(texto):
    '''
    Función que recibe la lista de registros y devuelve una lista 
    con todas las fechas de acuerdo a los rdiferentes formatos que se tengan
    '''
    # Vamos a generar expresiones regulares para cada formato que indentificamos:
    # Fecha xx/xx/xx | xx/xx/xxxx | x/x/xx
    regex1 = r"(0?[0-9]|[12][0-9])(/|-)(0?[1-9]|[1-2][0-9]|3[0-2])(/|-)((18|19|20)\d\d|\d\d)"
    # Mar-20-2009; Mar 20, 2009; March 20, 2009; Mar. 20, 2009; Mar 20 2009;
    regex2 = r"(?:Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:enver)?|Dec(?:ember)?)(?:|.?)( |-)([0-9][0-9]|0?[0-9])(?: |,|-?)(?: | ?)(((18|19|20)\d\d))"
    #12 March 2004; 21 Oct 1977
    regex3 = r"([0-9][0-9]|0?[0-9])(?:|.?)( |-)(?:Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:enver)?|Dec(?:ember)?)(?: |,|-?)(?: | ?)(((18|19|20)\d\d))"
    #Feb 2009; Sep 2009; Oct 2010
    regex4 = r"(Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:enver)?|Dec(?:ember))(?: |,|-?)(?: | ?)(((18|19|20)\d\d))"
    #6/2008; 12/2009
    regex5 = r"([0-9][0-9]|0?[0-9])/((18|19|20)\d\d|\d\d)"
    #2009; 2010
    regex6 = r"((18|19|20)\d\d)"

    # Generamos una lista con las expresiones regulares
    regex = [regex1, regex2, regex3, regex4, regex5, regex6]
    fechas = []

    # Vamos a iterar sobre la lista de resgitros
    for i in texto:
        j = 0
        k = 0
        lis = []
        # Vamos a correr un ciclo while hasta que se haga match con alguna fecha
        while k != 1:
            matches = re.finditer(regex[j], i)
            # Al hacer match vamos a obtener el resgistro y concatenarlo a 
            # una lista
            for matchNum, match in enumerate(matches, start=1):
                lis = [match.group()]
            k = len(lis)
            j += 1
            if j == 6:
                k = 1
        fechas = fechas + lis
    # Regresamos la lista con las fechas concatenadas
    return(fechas)
```

```python
# Aplicamos nuestra función a nuestros datos
fechas = fechas(contents)
```

```python
# Observamos la longitud 
len(fechas)
```

```python
# Vamos a transformar a fechas

fechas = pd.to_datetime(fechas)
```

```python
# Dado la ambiguedad de la base de datos a crear
# una funcion para aquellos valores fuera de rango:
import numpy as np

dates = np.where(fechas.year > 2022, fechas - pd.offsets.DateOffset(years=100), fechas)
```

```python
# Finalmmente Ordenammos
dates = sorted(dates)
dates
```
